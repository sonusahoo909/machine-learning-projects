```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprice.csv')
```

```python
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['area']] , df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

```python
import pickle
```

```python
with open('model_pickle', 'wb') as f:
    pickle.dump(model , f)
```

```python
with open('model_pickle' , 'rb') as f:
    mp = pickle.load(f)
```

```python
 mp.predict(5000) # error show kar rha hai but predict model
```
