```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

```

```python
df = pd.read_csv('homeprice.csv')
```

```python
df
```

```python
%matplotlib inline 
plt.xlabel('area(sqe ft)')
plt.ylabel('price(us$)')
plt.scatter(df.area,df.price,color = 'red' , marker='+')
```

```python
reg  = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]]) # ans = 628715.7534151643
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = m*x +b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline 
plt.xlabel('area(sqe ft)')
plt.ylabel('price(us$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
### linear regression by multiple variables
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df
```

```python
import math
df.bedroom.median()
```

```python
median_bedrooms = math.floor(df.bedroom.median())
```

```python
df.bedroom = df.bedroom.fillna(median_bedrooms)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedroom', 'age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```

```python
137.25*3000 + -26025*3 + -6825*40 + 383724.9999999998
```
