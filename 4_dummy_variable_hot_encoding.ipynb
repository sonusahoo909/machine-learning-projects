```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dummy variables.csv')
```

```python
df
```

```python
df.Houses.isnull()
```

```python
dummies = pd.get_dummies(df.Houses )
```

```python
dummies
```

```python
merged = pd.concat([df, dummies],axis ='columns')
merged
```

```python
last = merged.drop(['Houses','west windsor'], axis = 'columns')
last
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = last.drop('price', axis = 'columns')
x
```

```python
y = last.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0 , 1]])
```

```python
model.predict([[3400,0 , 0]])
```

```python
model.score(x, y)
```
